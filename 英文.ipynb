```python
import calendar

year = 2025

# 儲存所有 2025 年的星期五日期
friday_dates = []

for month in range(1, 13):
    # calendar.monthcalendar(year, month) 傳回一個列表的列表，代表月曆網格
    # 每個子列表代表一週，星期一為 0，星期日為 6
    month_calendar = calendar.monthcalendar(year, month)
    
    for week in month_calendar:
        # 檢查星期五 (索引為 4) 是否有日期
        if week[4] != 0: # 0 表示該天不在這個月
            # week[4] 是當月這個星期的星期五的日期
            # 格式：年-月-日
            date_str = f"{year}-{month:02d}-{week[4]:02d}"
            friday_dates.append(date_str)

# 輸出所有找到的星期五日期
print(f"--- 2025 年的所有星期五 ---")
for date in friday_dates:
    print(date)
    
```
