```python
import calendar

year = 2025

# 儲存所有 2025 年的星期一日期
monday_dates = []

for month in range(1, 13):
    # calendar.monthcalendar(year, month) 傳回一個列表的列表，代表月曆網格
    # 每個子列表代表一週，星期一為 0，星期日為 6
    month_calendar = calendar.monthcalendar(year, month)
    
    for week in month_calendar:
        # 檢查星期日 (索引為 6) 是否有日期
        if week[0] != 0: # 0 表示該天不在這個月
            # week[0] 是當月這個星期的星期日的日期
            # 格式：年-月-日
            date_str = f"{year}-{month:02d}-{week[0]:02d}"
            monday_dates.append(date_str)

# 輸出所有找到的星期日日期
print(f"--- 2025 年的所有星期一 ---")
for date in monday_dates:
    print(date)
```
